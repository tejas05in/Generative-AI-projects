```python
from llama_index.core import VectorStoreIndex , SimpleDirectoryReader , ServiceContext
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core.prompts.prompts import SimpleInputPrompt
```

```python
!mkdir Data
```

```python
documents = SimpleDirectoryReader("/content/Data").load_data()
```

```python
documents
```

```python
system_prompt = """ You are a Q&A assistant. Your goal is to answer questions as accurately
as possible based on the instructions and context provided. """

# This will wrap the defaults prompts that are internal to llama_index
query_wrapper_prompt = SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
!huggingface-cli login
```

```python
import torch

llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature":0.0,"do_sample":False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="mistralai/Mistral-7B-Instruct-v0.1", # change model_name and tokenizer name for different models
    model_name="mistralai/Mistral-7B-Instruct-v0.1",
    device_map="auto",
    # Uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype":torch.float16,"load_in_8bit":True}
)
```

```python
from langchain.embeddings.huggingface import HuggingFaceBgeEmbeddings
from llama_index.core.indices.service_context import ServiceContext
from llama_index.embeddings.langchain import LangchainEmbedding

embed_model = LangchainEmbedding(
    HuggingFaceBgeEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
)
```

```python
# service_context = ServiceContext.from_defaults(
#     chunk_size = 1024,
#     llm=llm,
#     embed_model=embed_model
# )
# Uncomment below code if above code is depricated
from llama_index.core import Settings
Settings.chunk_size = 1024
Settings.llm=llm
Settings.embed_model=embed_model
```

```python
index = VectorStoreIndex.from_documents(documents)
```

```python
print("Number of nodes:", len(index.docstore.docs))
```

```python
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("What is correlation?")
```
